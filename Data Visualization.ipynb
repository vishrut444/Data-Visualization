# DATA VISUALIZATION 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#so that all the graphs are displayed on the jupyter notebook itself
%matplotlib inline
```

```python
x = np.arange(0,11)
y = x**3
```

## 1) Line Charts

```python
plt.plot(x,y)
```

### Axis label, chart title

```python
plt.plot(x,y)
plt.title('My First chart')#title to chart
plt.xlabel('X-Axis')#name to x axis
plt.ylabel('Y-Axis')#name to y axis
#plt.show()-use this if you have not use %matplotlib inline above
```

### Multiple lines

```python
x = np.arange(0,11)
y = x**4

x2=np.arange(0,11)
y2=x**3

plt.plot(x,y,linewidth=3)#width of the line
plt.plot(x2,y2,linewidth=3)
plt.title('My second chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

### adding/edting color/legends to thecharts

```python
plt.plot(x,y,'r',label='First data',linewidth=3)#for red-r
plt.plot(x2,y2,'y',label='Second data',linewidth=3)#for yellow-y
plt.legend()#Rule whenever we use label wehave to use legend
#in this legend is at top left
# but we can put it any where 
```

### Adjusting the Size of the figure and baground grid and marker

```python
x = np.arange(0,11)
y = x**2

x2=np.arange(0,11)
y2=x**3

plt.figure(figsize=(10,10))# for adjusting size
plt.plot(x,y,linewidth=3,label='first line',marker='*',markersize=10)
plt.plot(x2,y2,linewidth=3,label='second line',marker='o',markersize=10)
plt.title('My second chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)# for adding grid
plt.legend(loc=0)
```

## 2) BAR Charts - (we create it whenever we have categorical variable and we want to check its value)

```python
x = ['India','USA','Japan','France','Israle']
y = [6,7,8,9,2]
plt.bar(x,y,label='Bar1',color='y')
plt.xlabel("countries")
plt.ylabel("Inflation rate in %")
plt.title("Bar Graph")
```

```python
x = ['India','USA','Japan','France','Israle']
y = [6,7,8,9,2]
plt.barh(x,y,label='Bar1',color='r')#for horizontal just add .barh instead of .bar
plt.xlabel("countries")
plt.ylabel("Inflation rate in %")
plt.title("Bar Graph")
```

### multiple bars

```python
x1 = ['India','USA','Japan','France','Israle']
y1 = [6,7,8,9,2]
x2 = ['India','USA','Japan','France','Israle']
y2 = [1,2,3,4,5]
plt.bar(x1,y1,label='Bar1',color='y')
plt.bar(x2,y2,label='Bar2',color='g')

plt.xlabel("countries")
plt.ylabel("Inflation rate in %")
plt.title("Bar Graph")
plt.legend()
```

## 3) Histograms

```python
plt.figure(figsize=(8,5))
stock_price = [12,33,44,54,65,76,45,34,55,33,23,23,33,43,43,54,67,89,67,87,56,88,45,65,31]
plt.hist(stock_price,bins=20)#more bind more no of individual value will occure
plt.xlabel('Stock Price')
plt.ylabel('Frequency')
plt.title("Histogram")
plt.grid(True)
```

## 5) PIE CHARTS

```python
raw_data = {'names':['vishrut','shreyash','yash','anand','ankit'],
           'jan_score':[128,127,126,125,124],
           'feb_score':[29,28,27,26,25],
           'march_score':[9,8,7,6,5]}
df = pd.DataFrame(raw_data,columns=['names','jan_score','feb_score','march_score'])
df['total_score'] = df['jan_score']+df['feb_score']+df['march_score']
df
```

```python
plt.pie(df['total_score'],labels=df['names'],autopct='%.2f%%')
plt.axis('equal')
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['total_score'],labels=df['names'],autopct='%.2f%%',explode=(0.5,0,0,0,0))#explodes tells us that by ho w nuch % our graaph must come out
plt.axis('equal')
plt.axis('equal')
plt.show()
```

## 6) SUBPLOTS

### Horizontal subplots

```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,5,4,1,4,5,7])

plt.subplot(1,2,1)#it means one rows 2 columns and last 1 stands for 1st chart
plt.plot(x,y,'r')
plt.title("Graph-1")

plt.subplot(1,2,2)
plt.plot(x,y,'m')
plt.title("Graph-2")

plt.tight_layout(w_pad=10)#it increases the space b/w two graphs
```

### Vertical Subplots

```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,5,4,1,4,5,7])

plt.subplot(2,1,1)#tworows,singlecolumn,1stgraph
plt.plot(x,y,'y')
plt.title("Graph-1")

plt.subplot(2,1,2)
plt.plot(x,y,'g')
plt.title("Graph-2")

plt.tight_layout(h_pad=10)#it increases the space b/w two graphs
```

### Subplot 2X2

```python
plt.figure(figsize=(20,15))
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,5,4,1,4,5,7])

plt.subplot(2,2,1)
plt.plot(x,y,'y')
plt.title("Graph-1")

plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.title("Graph-2")

plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.title("Graph-3")

plt.subplot(2,2,4)
plt.plot(x,y,'g')
plt.title("Graph-4")

plt.tight_layout(h_pad=15)
plt.tight_layout(w_pad=10)
```
