# DATA VISUALIZATION USING SEABORN 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_style('darkgrid')
```

```python
tips = sns.load_dataset('tips')#tips is an inbuilt dataset
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.describe(include='all')#it includes all categorical variables also
```

## 1) Dist Plot

```python
#it is use to show the distribution of univariant set of observations (one variable)
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)#remove kernal Density line
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=35)
```

## 2) Joint plot

```python
#it is use to show the distribution of bivariant set of observations(two variable)
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## 3) Pair Plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')#adds hue for categorical variable
plt.suptitle(t='My First PairPlot')
```

```python
sns.pairplot(tips, hue='day')
```

## PLOTTING CATEGORICAL DATA

### 4) BARPLOT 

```python
sns.barplot(x='sex',y='total_bill',data=tips)#this shows across mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)#this is about standard deviation
```

### 5) CountPlot

```python
sns.countplot(x='sex',data=tips)
```

### 6) Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### 7) Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### 8) HEATMAP

```python
#needs data to be in matrix form
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True)#for seenig values in the box we put annot=True
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')#for changing color of heatmap
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='Wistia')#for changing color of heatmap
```

### 9) FACEGRID

```python
sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```
